```python
# import basic libraries
import pandas as pd
import datetime
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```python
data = pd.read_csv('./ANZdataset.csv')
```

```python
salary_data = data[data["txn_description"]=="PAY/SALARY"].groupby(["customer_id"], as_index=False).sum()
salary_data.head()
```

```python
data[data["txn_description"]=="PAY/SALARY"].groupby(["customer_id"], as_index=False).count()
```

```python
salary_data['annual_salary'] = salary_data['amount'] * 4
salary_data.head()
```

```python
sns.distplot(salary_data.annual_salary)
plt.title('Histogram of Customer-Merchant Distance', fontsize=18)
plt.xlabel('Customer-Merchant Distance', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
```
